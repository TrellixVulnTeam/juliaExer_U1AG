```julia
using Plots
```

```julia
xs=linspace(-π,π,100)
ys=sin.(xs)
plot(xs,ys)
```

```julia
using PyCall
using PyPlot#
```

```julia
@pyimport numpy as np
@pyimport matplotlib.pyplot
xs=np.linspace(-np.pi,np.pi,100)
ys=np.sin(xs)
pyplot.plot(xs,ys)
pyplot.show()
```
