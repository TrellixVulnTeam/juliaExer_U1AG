```julia
using Plots
```

```julia
function create_gif()
    rw = [cumsum(randn(10^4)) for i in 1:3]
    @gif for θ in 0:3:357
        plot(rw[1], rw[2], rw[3], lw=0.3, size=(500, 500), camera=(θ, 60))
    end
end
```

```julia
gr()
create_gif()
```

```julia
plotlyjs()
create_gif()
```

```julia
pyplot()
create_gif()
```
